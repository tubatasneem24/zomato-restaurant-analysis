# Due to the large volume of data with hundreds of restaurant types, visualizing everything would lead to clutter and make interpretation difficult. To ensure clarity and focus, we chose to analyze and visualize only the top 50 most popular and bottom 50 least common restaurant types.


```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='rate', kde=True, bins=20, color='teal')
plt.title("Distribution of Ratings", fontsize=25)
plt.xlabel("Rating", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(4, 4))
sns.countplot(data=df, x='online_order', hue='online_order', palette="coolwarm")
plt.title("Online Order Availability", fontsize=25)
plt.xlabel("Online Order", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.show()
```

# The graph of online order availability shows that restaurants offering online ordering slightly outnumber those that do not. This indicates a growing preference for online ordering but also highlights that a significant number of establishments have yet to adopt it.

```python
plt.figure(figsize=(15, 15))
sns.countplot(data=df, y='area', order=df['area'].value_counts().index)
plt.title("Restaurant Concentration by Area", fontsize=40)
plt.xlabel("Count", fontsize=25)
plt.ylabel("Area", fontsize=25)
plt.show()

```

#  Areas with higher counts indicate densely populated or commercially active regions, while areas with lower counts reflect less developed or niche locations.

```python

plt.figure(figsize=(4, 4))
sns.countplot(data=df, x='online_order', hue='table booking', palette="YlGnBu")
plt.title("Online Order vs Table Booking", fontsize=20)
plt.xlabel("Online Order", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.legend(title="Table Booking", loc="upper right")
plt.show()
```

# Since the difference is minimal, both features appear important for customer convenience. Restaurants should aim to provide both online ordering and table booking options to cater to diverse customer preferences.

# Average Cost By Area⬇️

```python
area_cost = df.groupby('area')['avg_cost'].mean().sort_values()
area_cost.plot(kind='barh', figsize=(15, 15), color='darkblue')
plt.title("Average Cost for Two People by Area", fontsize=35)
plt.xlabel("Avg Cost", fontsize=25)
plt.ylabel("Area", fontsize=25)
plt.show()
```

# High-end areas show a greater average cost for dining, while more budget-friendly options are concentrated in mid-range or lower-cost areas.

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['rate', 'num_ratings', 'avg_cost']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation between Metrics", fontsize=30)
plt.show()
```

#  Higher-rated restaurants tend to attract slightly more customer ratings, indicating that quality influences engagement.

```python
# Top 10 popular areas
top_areas = df['area'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_areas.values, hue=top_areas.index)
plt.title("Top 10 Popular Areas", fontsize=30)
plt.xlabel("Count", fontsize=25)
plt.ylabel("Area", fontsize=25)
plt.show()

# Top 10 popular cuisines
top_cuisines = df['cuisines type'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, hue=top_cuisines, palette="mako")
plt.title("Top 10 Popular Cuisines", fontsize=30)
plt.xlabel("Count", fontsize=25)
plt.ylabel("Cuisine Type", fontsize=25)
plt.show()
```

# Byresandra, Tavarekere, and Madiwala dominate with the highest count, indicating that this area is a major hotspot for dining establishments.

# North Indian and Chinese cuisines emerge as the most common, followed by South Indian dishes. Fast food and multi-cuisine offerings (e.g., South Indian, North Indian, and Chinese) also contribute significantly to the diversity of available dining options.

```python
# Average cost by area
avg_cost_area = df.groupby('area')['avg_cost'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_cost_area.values, hue = avg_cost_area.index, palette="coolwarm")
plt.title("Top 10 Areas by Average Cost", fontsize=25)
plt.xlabel("Average Cost (for two people)", fontsize=25)
plt.ylabel("Area", fontsize=25)
plt.show()
```
